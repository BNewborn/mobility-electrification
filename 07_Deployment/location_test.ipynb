```python
def location_df(commuter_df):
    commuter_df['sequence']=commuter_df.groupby(['PUMAKEY_HOME']).cumcount()
    location_test['sequence']=location_test.groupby(['PUMAKEY_HOME']).cumcount()
    commuter_df = commuter_df.merge(right=location_test[['PUMAKEY_HOME','sequence','lat','lon']], on=["sequence","PUMAKEY_HOME"], how='left')
    return commuter_df
```

```python
commuter_model_0 = location_df(commuter_model_0)

commuter_model_1 = location_df(commuter_model_1)
commuter_model_2 = location_df(commuter_model_2)
commuter_model_3 = location_df(commuter_model_3)
commuter_model_4 = location_df(commuter_model_4)

commuter_model_5 = location_df(commuter_model_5)
commuter_model_6 = location_df(commuter_model_6)
commuter_model_7 = location_df(commuter_model_7)
commuter_model_8 = location_df(commuter_model_8)

commuter_model_9 = location_df(commuter_model_9)
commuter_model_10 = location_df(commuter_model_10)
commuter_model_11 = location_df(commuter_model_11)
commuter_model_12 = location_df(commuter_model_12)


commuter_model_0 = commuter_model_0.to_pickle(f"{save_dir}/Baseline_2019_Mode/{commuter_fname}")

commuter_model_1 = commuter_model_1.to_pickle(f"{save_dir}/HighWFH_Mix/{commuter_fname}")
commuter_model_2 = commuter_model_2.to_pickle(f"{save_dir}/HighWFH_Transit/{commuter_fname}")
commuter_model_3 = commuter_model_3.to_pickle(f"{save_dir}/HighWFH_Micro/{commuter_fname}")
commuter_model_4 = commuter_model_4.to_pickle(f"{save_dir}/HighWFH_Car/{commuter_fname}")

commuter_model_5 = commuter_model_5.to_pickle(f"{save_dir}/MidWFH_Mix/{commuter_fname}")
commuter_model_6 = commuter_model_6.to_pickle(f"{save_dir}/MidWFH_Transit/{commuter_fname}")
commuter_model_7 = commuter_model_7.to_pickle(f"{save_dir}/MidWFH_Micro/{commuter_fname}")
commuter_model_8 = commuter_model_8.to_pickle(f"{save_dir}/MidWFH_Car/{commuter_fname}")

commuter_model_9 = commuter_model_9.to_pickle(f"{save_dir}/NoWFH_Mix/{commuter_fname}")
commuter_model_10 = commuter_model_10.to_pickle(f"{save_dir}/NoWFH_Transit/{commuter_fname}")
commuter_model_11 = commuter_model_11.to_pickle(f"{save_dir}/NoWFH_Micro/{commuter_fname}")
commuter_model_12 = commuter_model_12.to_pickle(f"{save_dir}/NoWFH_Car/{commuter_fname}")
```
