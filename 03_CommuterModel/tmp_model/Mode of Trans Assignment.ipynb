```python
import pandas as pd
import numpy as np
import random
```

```python
sample = {'AUTO':[0, 0, 1, 1, 0], 'BUS':[0, 1, 0, 1, 0], 'BIKE':[0, 0, 1, 1, 0], 'WFH':[0,0,1,0,0]}
df = pd.DataFrame(data=sample)
df['Rand Assignment'] = ""
df
```

```python
modes = ['AUTO', 'BUS', 'BIKE', 'WFH']

for index, rows in df.iterrows():
    available_indices = []
    if df['AUTO'][index] == 1:
        available_indices.append(0)
    if df['BUS'][index] == 1:
        available_indices.append(1)
    if df['BIKE'][index] == 1:
        available_indices.append(2)
    if df['WFH'][index] == 1:
        available_indices.append(3)
    if len(available_indices) == 0:
        df['Rand Assignment'][index] = "No option"
    else:
        RandAssignment = random.choice(available_indices)
        df['Rand Assignment'][index] = modes[RandAssignment]
    
df
```
